```python
import pandas as pd
```

```python
people = {'first':['nishant', 'akshay', 'manoj'],
         'last':['chauhan', 'kumar', 'chauhan'],
         'email':['nich@gmail.com', 'akku@gmail.com', 'mach@gmail.com']}
df = pd.DataFrame(people)
df
```

```python
#using filter
filt = (df['last'] == 'chauhan')
df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
filt = (df['last'] == 'chauhan')
df[filt]['email'] # working same as .loc method
```

```python
#using and, or 
filt = (df['last'] == 'chauhan') & (df['email'] == 'mach@gmail.com')
df.loc[filt, 'email']
```

```python
#using and, or 
filt = (df['last'] == 'chauhan') | (df['email'] == 'mach@gmail.com')
df.loc[filt, 'email']
```

```python
#using reverse of filter
filt = (df['last'] == 'chauhan') | (df['email'] == 'mach@gmail.com')
df.loc[-filt, 'email']
```
