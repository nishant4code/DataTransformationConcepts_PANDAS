## Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
df.head()
```

### Task-1 Salary parsing

```python
df.shape
```

```python
(df['Salary Estimate']!= '-1').value_counts()
```

```python
#removing rows with 'Salary Estimate' == '-1', resetting the dataset
df = df[df['Salary Estimate']!= '-1']
df.head(2)
```

```python

salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary.unique()
```

```python
minus_kd = salary.apply(lambda x: x.replace('$', '').replace('K', ''))
minus_kd.unique()
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', ''))
min_hr.unique()
```

```python
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
```

```python
df['min_salary']
```

```python
df['min_salary'].dtype
```

```python
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df['max_salary']
```

```python
df['avg_salary'] = (df.min_salary+df.max_salary)/2
df['avg_salary']
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
# 24 Employees salary upon per hour
# 718 Employees salary upon monthly basis
df['hourly'].value_counts()
```

```python
df.head()
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df['employer_provided'].unique()
```

```python
df['employer_provided'].value_counts()
```

```python
df.head()
```

```python
df.head(2)
```

### company name text only

```python
#df['Company Name'].apply(lambda x: x.split('\n')[0])
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-4], axis = 1)
df['company_txt'].unique()
```

### state field

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
#df.head()
df.job_state.value_counts()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis = 1)
```

```python
df['same_state'].value_counts()
```

### age of company

```python
#df['Founded'].unique()
#filt = (df['Founded'] == -1)
#df.loc[filt, 'Founded'].value_counts()
df['age'] = df['Founded'].apply(lambda x: x if x<1 else 2022-x)
df['age']
```

### parsing of job description (python, etc)

```python
df.head(2)
```

```python
df['Job Description'][0]
```

```python
#Data Science tools
# Python, R Studio, Spark, AWS, excel
```

```python
# checking for Python
#df['python_yn']
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['python_yn'].value_counts()
```

```python
# checking for R Studio
df['python_r_studio'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['python_r_studio'].value_counts()
```

```python
# checking for Spark
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['spark'].value_counts()
```

```python
# checking for AWS
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['aws'].value_counts()
```

```python
# checking for excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['excel'].value_counts()
```
