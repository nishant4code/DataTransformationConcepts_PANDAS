```python
import pandas as pd
import numpy as np
```

```python
people = {'first':['nishant', 'akshay', 'manoj', 'shashank', np.nan, None, 'NA'],
         'last':['chauhan', 'kumar', 'chauhan', 'kumar', np.nan, np.nan, 'Missing'],
         'email':['nich@gmail.com', 'akku@gmail.com', 'mach@gmail.com', None, np.nan, 'Anonymous@gmail.com', 'NA'],
         'age':[32,33,45,38, None, None, 'Missing']}
df = pd.DataFrame(people)
df
```

```python
df.dropna(axis = 'index', how = 'all')
```

```python
df.dropna(axis = 'columns', how = 'any')
```

```python
#drop only those rows where email is missing by using 'subset'
df.dropna(axis = 'index', how = 'any', subset=['email'])
```

```python
#drop only those rows where email & last is missing by using 'subset'
df.dropna(axis = 'index', how = 'all', subset=['last', 'email'])
```

```python
#drop only those rows where email & last is missing by using 'subset'
df.dropna(axis = 'index', how = 'any', subset=['last', 'email'])
```

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
df.dropna(axis = 'index', how = 'all', subset = ['last', 'email'])
```
